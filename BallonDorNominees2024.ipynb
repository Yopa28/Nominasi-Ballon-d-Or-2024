```python
# Membuat DataFrame
df = pd.DataFrame(data)
```

```python
# Analisis Distribusi Posisi
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Posisi', palette='Set2', order=df['Posisi'].value_counts().index)
plt.title('Distribusi Posisi Pemain', fontsize=16)
plt.xlabel('Posisi', fontsize=12)
plt.ylabel('Jumlah Pemain', fontsize=12)
plt.show()
```

```python
# Analisis Distribusi Usia
plt.figure(figsize=(8, 6))
sns.histplot(df['Usia'], bins=10, kde=True, color='skyblue')
plt.title('Distribusi Usia Pemain', fontsize=16)
plt.xlabel('Usia', fontsize=12)
plt.ylabel('Jumlah Pemain', fontsize=12)
plt.show()
```

```python
# Analisis Negara Asal
plt.figure(figsize=(10, 6))
df['Negara'].value_counts().head(10).plot(kind='bar', color='teal')
plt.title('10 Negara Teratas dalam Nominasi', fontsize=16)
plt.xlabel('Negara', fontsize=12)
plt.ylabel('Jumlah Pemain', fontsize=12)
plt.show()
```

```python
# Analisis Rata-rata Usia Berdasarkan Posisi
avg_age_by_position = df.groupby('Posisi')['Usia'].mean().sort_values()
plt.figure(figsize=(8, 6))
avg_age_by_position.plot(kind='barh', color='coral')
plt.title('Rata-rata Usia Berdasarkan Posisi', fontsize=16)
plt.xlabel('Rata-rata Usia', fontsize=12)
plt.ylabel('Posisi', fontsize=12)
plt.show()
```
